```python
avg_purchase = df['Purchase_Amount'].mean()
df['Purchase_Amount'].fillna(avg_purchase, inplace=True)
```

```python
df['City'].isnull().sum()
```

```python
df.dropna(subset=['Age'], inplace=True)
```

```python
df['Gender'] = df['Gender'].str.upper()

df['Gender'] = df['Gender'].replace({
    'M': 'Male',
    'F': 'Female'
})
```

```python
df['City'] = df['City'].str.upper().str.strip()
```

```python
df['Country'] = 'India'
```

```python
df['Signup_Date'] = pd.to_datetime(df['Signup_Date'])
```

```python
df['Last_Purchase_Date'] = pd.to_datetime(df['Last_Purchase_Date'])
```

```python
df['Purchase_Amount'].mean()
```

```python
df.groupby('City')['Purchase_Amount'].mean()
```

```python
df['City'].value_counts()
```

```python
overall_avg = df['Purchase_Amount'].mean()
df[df['Purchase_Amount'] > overall_avg]
```

```python
overall_avg = df['Purchase_Amount'].mean()
df['Customer_Type'] = df['Purchase_Amount'].apply(
    lambda x: 'High Value' if x > overall_avg else 'Regular'
)
```
